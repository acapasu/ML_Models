# Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,30)
y = x**2
```

```python
plt.plot(x,y)
```

```python
plt.plot(x, y) 
plt.xlabel('Titulo del eje X')
plt.ylabel('Titulo del eje Y')
plt.title('Titulo del grafico')
```

```python
plt.plot(x, y, 'red')
plt.xlabel('Titulo del eje X')
plt.ylabel('Titulo del eje Y')
plt.title('Titulo del grafico')
plt.show()
```

```python
plt.plot(x, y, 'red')
plt.plot(y,x,'blue')
```

```python
plt.subplot(1,2,1)
plt.plot(x, y, 'red')
plt.subplot(1,2,2)
plt.plot(y,x,'blue')
```

```python
fig = plt.figure()
ejes = fig.add_axes([0, 0, 1, 1])

ejes.plot(x, y, 'blue')
ejes.set_xlabel('Titulo del eje X')
ejes.set_ylabel('Titulo del eje Y')
ejes.set_title('Titulo del grafico')
```

```python
ejes
```

```python
fig = plt.figure()

ejes1 = fig.add_axes([0, 0, 1, 1])
ejes2 = fig.add_axes([0.15, 0.6, 0.4, 0.3])


ejes1.plot(x, y, 'blue')
ejes1.set_xlabel('eje 1 X')
ejes1.set_ylabel('eje 1 Y')
ejes1.set_title('Titulo primer eje')


ejes2.plot(y, x, 'red')
ejes2.set_xlabel('eje 2 X')
ejes2.set_ylabel('eje 2 Y')
ejes2.set_title('Titulo eje 2');
```

```python
fig, ejes = plt.subplots()

ejes.plot(x, y, 'red')
ejes.set_xlabel('x')
ejes.set_ylabel('y')
ejes.set_title('Titulo');
```

```python
ejes
```

```python
fig, ejes = plt.subplots(nrows=1, ncols=2)
```

```python
ejes
```

```python
for e in ejes:
    e.plot(x,y, 'b')
    e.set_xlabel('eje x')
    e.set_ylabel('eje y')
    e.set_title('Titulo')

fig
```

## Texto sobre graficos

```python
fig = plt.figure()

eje = fig.add_axes([0,0,1,1])

eje.plot(x, x**2, label="x**2")
eje.plot(x, x**3, label="x**3")
eje.legend()
```

```python
fig = plt.figure()

eje = fig.add_axes([0,0,1,1])

eje.plot(x, x**2, label="x**2")
eje.plot(x, x**3, label="x**3")
eje.legend(loc = 5)
```

## Estetica

```python
fig, eje = plt.subplots()
eje.plot(x, x**2, 'b--')
eje.plot(x, x**3, 'g.-')
```

```python
fig, eje = plt.subplots()
eje.plot(x, x**2, 'b--')
eje.plot(x, x**3, 'g.-')
eje.grid(color='b', alpha=0.5, linestyle='dashed', linewidth=0.5)

```

```python
fig, eje = plt.subplots()

eje.plot(x, x**2, color="blue", alpha=0.5)
eje.plot(x, x**3, color="#8B008B")
eje.plot(x, x**4, color="#FF8C00")
```

```python
fig, eje = plt.subplots(figsize=(12,6))

eje.plot(x, x+1, color="r", linewidth=0.25)
eje.plot(x, x+2, color="r", linewidth=0.50)
eje.plot(x, x+3, color="r", linewidth=1.00)
eje.plot(x, x+4, color="r", linewidth=2.00)
```

```python
fig, eje = plt.subplots()
eje.plot(x, x+5, color="g", lw=3, linestyle='-')
eje.plot(x, x+6, color="g", lw=3, ls='-.')
eje.plot(x, x+7, color="g", lw=3, ls=':')
```

```python
fig, eje = plt.subplots()
eje.plot(x, x**2, color="b", lw=2)[0].set_dashes([5, 10, 15, 10])
```

```python
fig, eje = plt.subplots()
eje.plot(x, x, color="b", lw=1, ls='-', marker='+')
eje.plot(x, x+1, color="b", lw=1, ls='-', marker='o')
eje.plot(x, x+2, color="b", lw=1, ls='-', marker='s')
eje.plot(x, x+3, color="b", lw=1, ls='-', marker='*')
```

```python
fig, eje = plt.subplots(figsize=(12,8))
eje.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
eje.plot(x, x+14, color="purple", lw=1, ls='-', marker='o', markersize=4)
eje.plot(x, x+15, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
eje.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', markersize=8, 
        markerfacecolor="y", markeredgewidth=3, markeredgecolor="green");
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].plot(x, x**2)
axes[0].set_title("por defecto")

axes[1].plot(x, x**2)
axes[1].set_ylim([0, 60])
axes[1].set_xlim([2, 5])
axes[1].set_title("como parametro");
```
