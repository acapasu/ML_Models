# PCA

## Import Libraries

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

from mpl_toolkits import mplot3d
```

## Import dataset

```python
cancer = load_breast_cancer()
df = pd.DataFrame(data=cancer.data, columns=cancer.feature_names)
df['target'] = pd.Series(cancer.target)
df_base = df.drop('target', axis = 1)
```

```python
df.head()
```

## Scale data

```python
scaler = StandardScaler()
scaled_X = scaler.fit_transform(df_base)
```

```python
scaled_X
```

## Scikit learn

```python
pca = PCA(n_components=2)
```

```python
principal_components = pca.fit_transform(scaled_X)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(principal_components[:,0],principal_components[:,1])
plt.xlabel('First principal component')
plt.ylabel('Second Principal Component')
```

```python
plt.figure(figsize=(8,6))
plt.scatter(principal_components[:,0],principal_components[:,1],c=df['target'])
plt.xlabel('First principal component')
plt.ylabel('Second Principal Component')
```

## Aditional information

```python
df_comp = pd.DataFrame(pca.components_,index=['PC1','PC2'],columns=df_base.columns)
```

```python
df_comp
```

```python
plt.figure(figsize=(20,3),dpi=150)
sns.heatmap(df_comp,annot=True)
```

```python
pca.explained_variance_ratio_
```

```python
np.sum(pca.explained_variance_ratio_)
```

```python
explained_variance = []

for n in range(1,30):
    pca = PCA(n_components=n)
    pca.fit(scaled_X)
    
    explained_variance.append(np.sum(pca.explained_variance_ratio_))
```

```python
plt.plot(range(1,30),explained_variance)
plt.xlabel("Number of Components")
plt.ylabel("Variance Explained");
```

## 3D

```python
pca = PCA(n_components=3)
```

```python
principal_components = pca.fit_transform(scaled_X)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(principal_components[:,0],principal_components[:,1])
plt.xlabel('First principal component')
plt.ylabel('Second Principal Component')
```

```python
plt.figure(figsize=(8,8),dpi=150)
ax = plt.axes(projection='3d')
ax.scatter3D(principal_components[:,0],principal_components[:,1],principal_components[:,2],c=df['target']);
```
